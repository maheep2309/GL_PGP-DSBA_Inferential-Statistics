## **Problem 2**
The breaking strength of gunny bags used for packaging cement is normally distributed with a mean of 5 kg per sq. centimeter and a standard deviation of 1.5 kg per sq. centimeter. The quality team of the cement company wants to know the following about the packaging material to better understand wastage or pilferage within the supply chain. Answer the questions below based on the given information Also, provide an appropriate visual representation of your answers.

2.1 What proportion of the gunny bags have a breaking strength of less than 3.17 kg per sq cm?

2.2 What proportion of the gunny bags have a breaking strength of at least 3.6 kg per sq cm.?

2.3 What proportion of the gunny bags have a breaking strength between 5 and 5.5 kg per sq cm.?

2.4 What proportion of the gunny bags have a breaking strength NOT between 3 and 7.5 kg per sq cm.?

## **Solution 2**

```python
# Libraries to help with reading and manipulating data
import numpy as np
import pandas as pd

# Libraries to help with data visualization
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# Library to help with statistical analysis
import scipy as scy
import scipy.stats as stats
```

### 2.1 What proportion of the gunny bags have a breaking strength less than 3.17 kg per sq cm?

```python
# Mean breaking strength & Standard deviation given (in kg per sq cm)
mean_strength = 5
std_dev_strength = 1.5

# compute z-stat
z1 = (3.17 - mean_strength) / std_dev_strength
prob_z1 = round(stats.norm.cdf(z1),4)
print('Z Value:',z1)

#Create PDF Plot
density_fn = pd.DataFrame()
density_fn["x"] = np.linspace(mean_strength - 4*std_dev_strength, mean_strength + 4*std_dev_strength, 1000)
density_fn["pdf"] = stats.norm.pdf(density_fn["x"], mean_strength, std_dev_strength)
fig, ax = plt.subplots()
ax.plot(density_fn["x"], density_fn["pdf"])
plt.axvline(x=3.17, c="red")
fill_array1 = np.linspace(density_fn["x"].min(), 3.17, 1000)
plt.fill_between(fill_array1, stats.norm.pdf(fill_array1, mean_strength, std_dev_strength), color="red", label = 'P(X<3.17)')
plt.title('Normal Distribution of Gunny Bags Breaking-Strength')
plt.xlabel('Breaking Strength (kg per sq cm)')
plt.ylabel('Probability Density')
plt.legend()
plt.show()

print(prob_z1,'proportion (i.e.',prob_z1*100,'%) of the gunny bags have a breaking strength less than 3.17 kg per sq cm')
```

### 2.2 What proportion of the gunny bags have a breaking strength at least 3.6 kg per sq cm.?

```python
# Mean breaking strength & Standard deviation given (in kg per sq cm)
mean_strength = 5
std_dev_strength = 1.5

# compute z-stat
z2 = (3.6 - mean_strength) / std_dev_strength
prob_z2 = round(1 - stats.norm.cdf(z2),4)
print('Z Value:',z2)

#Create PDF Plot
density_fn = pd.DataFrame()
density_fn["x"] = np.linspace(mean_strength - 4*std_dev_strength, mean_strength + 4*std_dev_strength, 1000)
density_fn["pdf"] = stats.norm.pdf(density_fn["x"], mean_strength, std_dev_strength)
fig, ax = plt.subplots()
ax.plot(density_fn["x"], density_fn["pdf"])
plt.axvline(x=3.6, c="green")
fill_array2 = np.linspace(3.6, density_fn["x"].max(), 1000)
plt.fill_between(fill_array2, stats.norm.pdf(fill_array2, mean_strength, std_dev_strength), color="green", label='P(X>=3.6)')
plt.title('Normal Distribution of Gunny Bags Breaking-Strength')
plt.xlabel('Breaking Strength (kg per sq cm)')
plt.ylabel('Probability Density')
plt.legend()
plt.show()

print(prob_z2,'proportion (i.e.',prob_z2*100,'%) of the gunny bags have a breaking strength at least 3.6 kg per sq cm')
```

### 2.3 What proportion of the gunny bags have a breaking strength between 5 and 5.5 kg per sq cm.?

```python
# Mean breaking strength & Standard deviation given (in kg per sq cm)
mean_strength = 5
std_dev_strength = 1.5

# compute z-stat
z3_UL = (5.5 - mean_strength) / std_dev_strength
z3_LL = (5 - mean_strength) / std_dev_strength
prob_z3 = round(stats.norm.cdf(z3_UL) - stats.norm.cdf(z3_LL),4)
print('Z Value_Upper:',z3_UL)
print('Z Value_Lower:',z3_LL)

#Create PDF Plot
density_fn = pd.DataFrame()
density_fn["x"] = np.linspace(mean_strength - 4*std_dev_strength, mean_strength + 4*std_dev_strength, 1000)
density_fn["pdf"] = stats.norm.pdf(density_fn["x"], mean_strength, std_dev_strength)
fig, ax = plt.subplots()
ax.plot(density_fn["x"], density_fn["pdf"])
plt.axvline(x=5, c="orange")
plt.axvline(x=5.5, c="orange")
fill_array3 = np.linspace(5, 5.5, 1000)
plt.fill_between(fill_array3, stats.norm.pdf(fill_array3, mean_strength, std_dev_strength), color="orange", label='P(5<X<5.5)')
plt.title('Normal Distribution of Gunny Bags Breaking-Strength')
plt.xlabel('Breaking Strength (kg per sq cm)')
plt.ylabel('Probability Density')
plt.legend()
plt.show()

print(prob_z3,'proportion (i.e.',round(prob_z3*100,2),'%) of the gunny bags have a breaking strength between 5 and 5.5 kg per sq cm')
```

### 2.4 What proportion of the gunny bags have a breaking strength NOT between 3 and 7.5 kg per sq cm.?

```python
# Mean breaking strength & Standard deviation given (in kg per sq cm)
mean_strength = 5
std_dev_strength = 1.5

# compute z-stat
z3_n_UL = (7.5 - mean_strength) / std_dev_strength
z3_n_LL = (3 - mean_strength) / std_dev_strength
prob_z4 = round(stats.norm.cdf(z3_n_LL) + (1 - stats.norm.cdf(z3_n_UL)),4)
print('Z Value_Upper:',z3_n_UL)
print('Z Value_Lower:',z3_n_LL)

#Create PDF Plot
density_fn = pd.DataFrame()
density_fn["x"] = np.linspace(mean_strength - 4*std_dev_strength, mean_strength + 4*std_dev_strength, 1000)
density_fn["pdf"] = stats.norm.pdf(density_fn["x"], mean_strength, std_dev_strength)
fig, ax = plt.subplots()
ax.plot(density_fn["x"], density_fn["pdf"])
plt.axvline(x=3, c="purple")
plt.axvline(x=7.5, c="purple")
fill_array4a = np.linspace(density_fn["x"].min(), 3, 1000)
fill_array4b = np.linspace(7.5, density_fn["x"].max(), 1000)
plt.fill_between(fill_array4a, stats.norm.pdf(fill_array4a, mean_strength, std_dev_strength), color="purple", label='P(X<3 OR X>7.5)')
plt.fill_between(fill_array4b, stats.norm.pdf(fill_array4b, mean_strength, std_dev_strength), color="purple")
plt.title('Normal Distribution of Gunny Bags Breaking-Strength')
plt.xlabel('Breaking Strength (kg per sq cm)')
plt.ylabel('Probability Density')
plt.legend()
plt.show()

print(prob_z4,'proportion (i.e.',round(prob_z4*100,2),'%) of the gunny bags have a breaking strength NOT between 3 and 7.5 kg per sq cm')
```

### 3.1 Zingaro has reason to believe that the unpolished stones may not be suitable for printing. Do you think Zingaro is justified in thinking so? -  State the null and alternate hypothese.  Conduct the hypothesis test and compute the p-value.  Write down conclusions from the test result.  Consider the level of significance as 5%"

```python
# Perform Visual Anlysis - Lets look at the Unpolished data for initial impressions
plt.figure(figsize=(10, 6))
plt.title('Histogram - Unpolished')
plt.xlabel('Unpolished')
plt.ylabel('Count')
sns.histplot(data=df_Zingaro,x=df_unpolished,kde=True)
plt.axvline(x=150, color='red', linestyle='--', label='Required Hardness Threshold (150)')
plt.show()

# The histogram above shows the distribution of the Brinell hardness values for unpolished stones.
# The red dashed line at 150 represents the required hardness threshold for printing.
```

```python
# Step 1: Define the null and alternate hypotheses:-
# Null Hypothesis (Hâ‚€): Mean hardness of unpolished stones is greater than or equal to 150 (ðœ‡ â‰¥ 150)
# Alternative Hypothesis (Hâ‚): Mean hardness of unpolished stones is less than 150 (ðœ‡ < 150)

#Step 2: Select Appropriate test:-
# We can perform one-sample t-test (single-tailed) to check if the mean Brinell hardness of the unpolished stones is significantly less than 150.

#Step 3: Decide the significance level:-
# It is given in the question to consider significance level as 5%

#Step 4: Prepare Data for the test:-
df_unpolished = df_Zingaro[col[0]]
mean_unpolished = 150

#Step 5: Calculate the test statistic and p-value
from scipy.stats import ttest_1samp #import the required functions
test_stat, p_value = ttest_1samp(df_unpolished, popmean = mean_unpolished, alternative = 'less')
print('The p-value is ', p_value)

#Step 6: Conclusion based on p-value
if p_value < 0.05:
    print(f'As the p-value {p_value} is less than the level of significance, we reject the null hypothesis.')
else:
    print(f'As the p-value {p_value} is greater than the level of significance, we fail to reject the null hypothesis.')

# Mean hardness of unpolished stones is less than 150 (ðœ‡ < 150). 
# There is sufficient evidence to support the claim that the unpolished stones may not suitable for printing.
```

### 3.2 Is the mean hardness of the polished and unpolished stones the same? -  State the null and alternate hypotheses. Conduct the hypothesis test. Write down conclusions from the test results. Consider the level of significance as 5%.

```python
# Perform Visual Anlysis
df_unpolished = df_Zingaro[col[0]]
df_polished = df_Zingaro[col[1]]

plt.figure(figsize=(10, 5))
sns.histplot(data=df_Zingaro,x=df_unpolished,kde=True, label='Unpolished Stones')
sns.histplot(data=df_Zingaro,x=df_polished,kde=True, label='Polished Stones')
plt.axvline(x=150, color='green', linestyle='--', label='Required Hardness Threshold (150)')
plt.xlabel('Brinell Hardness Index')
plt.ylabel('Frequency')
plt.title('Distribution of Hardness Values for Polished and Unpolished Stones')
plt.legend()
plt.show()

# Lets find the sample means and sample standard deviations for polished and unpolishedthe two samples
print('The mean hardness of polished stones is ' + str(round(df_Zingaro[col[1]].mean(),2)))
print('The mean hardness of unpolished stones is ' + str(round(df_Zingaro[col[0]].mean(),2)))
print('The standard deviation of polished stones is ' + str(round(df_Zingaro[col[1]].std(), 2)))
print('The standard deviation of unpolished stones is ' + str(round(df_Zingaro[col[0]].std(), 2)))
```

```python
# Step 1: Define the null and alternate hypotheses:-
# Null Hypothesis (Hâ‚€): The mean hardness of polished stones is equal to that of unpolished stones (i.e., Î¼â‚ = Î¼â‚‚)
# Alternative Hypothesis (Hâ‚): The mean hardness of polished stones is different from that of unpolished stones (i.e., Î¼â‚ â‰  Î¼â‚‚)

#Step 2: Select Appropriate test:-
# We can perform an independent two-sample t-test between the 2 means (for unequal standard deviations, as shown above)

#Step 3: Decide the significance level:-
# It is given in the question to consider significance level as 5%

#Step 4: Prepare Data for the test:-
df_unpolished = df_Zingaro[col[0]]
df_polished = df_Zingaro[col[1]]

#Step 5: Calculate the test statistic and p-value
from scipy.stats import ttest_ind #import the required functions
test_stat, p_value = ttest_ind(df_unpolished, df_polished, equal_var = False, alternative = 'two-sided')
print('The p-value is ', p_value)

#Step 6: Conclusion based on p-value
if p_value < 0.05:
    print(f'As the p-value {p_value} is less than the level of significance, we reject the null hypothesis.')
else:
    print(f'As the p-value {p_value} is greater than the level of significance, we fail to reject the null hypothesis.')

# This indicates that there is a significant difference in the mean hardness between polished and unpolished stones
```
